# HRF Gamma

```python
import scipy.stats as stats
import numpy as np
import seaborn as sns
delay = 6
loc=0
scale=3
time_stamps = np.arange(0,45,0.1)
response_function = stats.gamma.pdf(time_stamps, delay, loc=loc, scale=scale)
```

```python
sns.relplot(x=time_stamps, y=response_function)
```

```python
def downsampling(ds):
    sig  = []
    for s,e in zip(np.arange(0,ds.shape[0],15), np.arange(10,ds.shape[0],15)):
        sig.append(ds[s:e].mean() )
    return np.asarray(sig)* 1000000 #/ np.asarray(sig).sum()
```

```python
len(downsampling(response_function[:len(time_stamps)])), len(time_stamps),len(response_function)
```

```python
min(downsampling(response_function[:len(time_stamps)]))
```

```python
sns.relplot(downsampling(response_function[:len(time_stamps)]))
```
